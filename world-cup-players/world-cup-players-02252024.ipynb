```python
import pandas as pd
import numpy as np
import fnmatch as fnm

df_players = pd.read_csv('C:\\Users\\luis1\\OneDrive\\Documents\\Learning\\WorldCupPlayers.csv')

df_players.tail(5)
```

```python
# df_matches['Referee'].str.split(pat ='(').str[0]
# df_players[[[[[[['Event1', 'Event2', 'Event3', 'Event4', 'Event5', 'Event6']]]]]]] = 

player_expand = df_players['Event'].str.split(" ", expand=True)

df_players = pd.concat([df_players, player_expand], axis=1)

df_players
```

```python
#use wildcard to make string that starts with G to Goal
def G_check(row):
    if row['Event'].str.startswith('G'):
        return 'Goal'
    else: ''

df_players['Goal_String'] = df_players.apply(G_check, axis=0)

df_players
```

```python
#player goal check condition
#player_goal_check = df_players['0'].str.startswith('G', na=False)
def goal_0():
    if df_players['0'].str.startswith('G'):
        return True
    else: False
df_players['0'] = goal_0

def goal_1():
    if df_players['1'].str.startswith('G'):
        return True
    else: False
df_players['1'] = goal_1

def goal_2():
    if df_players['2'].str.startswith('G'):
        return True
    else: False
df_players['2'] = goal_2

def goal_3():
    if df_players['3'].str.startswith('G'):
        return True
    else: False
df_players['3'] = goal_3

def goal_4():
    if df_players['4'].str.startswith('G'):
        return True
    else: False
df_players['4'] = goal_4
```
